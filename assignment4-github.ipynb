```python
import sqlite3

# Connect to the SQLite database file
conn = sqlite3.connect('/kaggle/working/your_database_file.db')

# Create a cursor object to execute SQL queries
cursor = conn.cursor()

# Create the customers table
cursor.execute('''
    CREATE TABLE employees (
        year INTEGER PRIMARY KEY,
        number INTEGER,
    )
''')

# Commit the changes to the database
conn.commit()

# Insert data into the customers table
cursor.execute('''
    INSERT INTO employees (year, count)
    VALUES
        (2017, 700),
        (2019, 1019),
        (2020, 1677)
        (2021,3100),
        (2022, 5000),
''')

# Commit the changes to the database
conn.commit()

# Count the number of entries in the customers table
cursor.execute('SELECT 
    ((e1.count - e2.count) / e2.count) * 100 AS percentage_difference
FROM
    employees e1')
entry_count = cursor.fetchone()[0]

# Print the result
print("Percentage increase in employees over the last two years:", entry_count)

# Close the database connection
conn.close()
```
